```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile

from IPython.display import Image, Audio

font_size = 25
plt.rcParams.update({'font.size': font_size})
```

# 1.0 Reading in .wav files

```python
fs, signal_violin = wavfile.read("data_bhb/violin_fs=16kHz.wav")
```

```python
signal_violin = signal_violin.astype(np.float64)
t = np.arange(0, len(signal_violin)) * (1/fs)

plt.figure(figsize=(14,6))
plt.plot(t, signal_violin)
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude (Arbitrary Unit)')
plt.xlim(0)
plt.grid()
plt.show()
```

```python
Audio(signal_violin, rate=fs)
```

```python
signal_violin_abs = np.abs(signal_violin)

plt.figure(figsize=(14,6))
plt.plot(t, signal_violin_abs)
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude (Arbitrary Unit)')
plt.xlim(0)
plt.grid()
plt.show()
```

```python
Audio(signal_violin_abs, rate=fs)
```
